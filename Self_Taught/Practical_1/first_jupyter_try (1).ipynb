```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('3.2 cost_revenue_clean.csv')
```

```python
data
```

```python
X = data['production_budget_usd']
y = data['worldwide_gross_usd']
```

```python
plt.figure(figsize=(10,6))

plt.scatter(X, y, alpha = 0.3)
plt.title('Film Cost Vs Global Revenue',color ='red')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0,3000000000)
plt.xlim(0)

plt.show()
```

```python
X = pd.DataFrame(data,columns=['production_budget_usd'])
y = pd.DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
regression = LinearRegression()
```

```python
regression.fit(X, y)
```

Slope coefficient 

```python
## The slope coefficient tells how much the feature and label are related

regression.coef_
```

Intercept

```python
regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha = 0.3)
plt.plot(X, regression.predict(X),color='red',linewidth =6)
plt.title('Film Cost Vs Global Revenue',color ='red')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0,3000000000)
plt.xlim(0)

plt.show()
```

```python
### Goodness of fit (R2 or r2)/ it tells how much good our model is

regression.score(X,y)
```

The above 0.54 explain the variable of the production budget and the gross revenue and we can say that is very good .
